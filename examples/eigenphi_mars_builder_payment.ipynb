```python
from data_prep_utils import *
from plot_prep_utils import *
import pandas as pd

eigenphi_data = pd.read_csv('eigenphi_march_block_builder_rewards.csv')    
# print(eigenphi_data['builder_reward'].quantile([0.25, 0.5, 0.75, 0.95, 1.0]))

# 对 'builder_reward' 列进行排序，并取前50个最高值
top_50_rewards = eigenphi_data['builder_reward'].nlargest(50)

# 输出这50个值
print(top_50_rewards)


summary_stats = pd.Series({
    'Mean': eigenphi_data['builder_reward'].mean(),
    'Median': eigenphi_data['builder_reward'].median(),
    'Standard Deviation': eigenphi_data['builder_reward'].std(),
    'Skewness': eigenphi_data['builder_reward'].skew(),
    'Kurtosis': eigenphi_data['builder_reward'].kurt(),
    'Minimum': eigenphi_data['builder_reward'].min(),
    'Maximum': eigenphi_data['builder_reward'].max(),
    '25th Percentile': eigenphi_data['builder_reward'].quantile(0.25),
    '75th Percentile': eigenphi_data['builder_reward'].quantile(0.75),
    '99th Percentile': eigenphi_data['builder_reward'].quantile(0.99),
    'Total Count': eigenphi_data['builder_reward'].count(),
    'Number of Unique Values': eigenphi_data['builder_reward'].nunique()
})
print(summary_stats)
```
